```python
# Solving XOR problems with a Multi-Layer Perceptron (MLP) using TensorFlow and Keras
# import necessary libraries
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Define the XOR dataset
def create_xor_dataset():
    # Input data for XOR
    X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    # Output data for XOR
    y = np.array([[0], [1], [1], [0]])
    return X, y


# Create the XOR dataset
X, y = create_xor_dataset()


# Define the model architecture
def create_model():
    model = keras.Sequential(
        [
            layers.Dense(
                4, activation="relu", input_shape=(2,)
            ),  # Hidden layer with 4 neurons
            layers.Dense(1, activation="sigmoid"),  # Output layer with 1 neuron
        ]
    )
    return model
```

```python
# Create the model
model = create_model()
```

```python
# Compile the model
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
# Train the model
model.fit(X, y, epochs=1000)
```

```python
# After training, evaluate the model
test_loss, test_acc = model.evaluate(X, y)
print("Test accuracy:", test_acc)
```

```python
# Test the model with the XOR inputs
X_test = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_pred = model.predict(X_test)
for i, pred in enumerate(y_pred):
    print(f"Input: {X_test[i]}, Predicted Output: {np.round(pred).astype(int)[0]}")
```
