```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from scipy.signal import convolve2d
```

```python
# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# print matrix of the first image
print(X_train[0])

# plot the first image with its label
plt.imshow(X_train[0], cmap="gray")
plt.title("Label: " + str(y_train[0]))
```

```python
# Define vertical, horizontal and corner filters
vertical_filter = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
horizontal_filter = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
corner_filter = np.array([[1, 0, -1], [0, 0, 0], [-1, 0, 1]])
print(vertical_filter)
print(horizontal_filter)
print(corner_filter)
```

```python
# Apply filters to the first image
vertical_result = convolve2d(X_train[0], vertical_filter, mode="same")
horizontal_result = convolve2d(X_train[0], horizontal_filter, mode="same")
corner_result = convolve2d(X_train[0], corner_filter, mode="same")
```

```python
# plot the results
fig, ax = plt.subplots(1, 4, figsize=(15, 5))
ax[0].imshow(vertical_result, cmap="gray")
ax[0].set_title("Vertical Filter")
ax[1].imshow(horizontal_result, cmap="gray")
ax[1].set_title("Horizontal Filter")
ax[2].imshow(corner_result, cmap="gray")
ax[2].set_title("Corner Filter")
ax[3].imshow(X_train[0], cmap="grey")
ax[3].set_title("Original Image")
plt.show()
```
